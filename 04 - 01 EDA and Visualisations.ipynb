### Boolean indexing

The main difference between ‘and’ and ‘&’ in Python is:
and is a Logical AND that returns True if both the operands are true whereas ‘&’ is a bitwise operator in Python that acts on bits and performs bit by bit operation.
When an integer value is 0, it is considered as False when using logically.

source: geekforgeeks

```python
a, b = 9, 10
print(a & b)#line 1 
print(a and b)#line 2 
```

The first line is performing bitwise AND on a and b and the second line is evaluating the statement inside print and printing answer.
In line 1, a = 1001 (in binary), b = 1010, Performing & on a and b, gives us 1000 which is the binary value of decimal value __8__.
In line 2, the expression ‘a and b’ first evaluates a; if a is False (or Zero), its value is returned immediately which is __10__

## Computations on Groupby 
__size()__:
Number of rows in each group as a Series if as_index is True or a DataFrame if as_index is False.

```python
odf.groupby('Edition').size()
```

__agg([...])__:
Aggregate using one or more operations over the specified axis.

```python
odf.groupby(['Edition','NOC','Medal']).agg(['min','max','count'])

# the min and max work on numeric and integer values
```

```python
odf.groupby(['Edition','NOC','Medal']).agg(['count'])
```

### Show a summary of the total medals won by participating countries in the 2008 Olympics.

```python
lo = odf[odf.Edition == 2008]
lo
```

```python
#first groupby participating countries using the NOC and unstack by medals
g1 = lo.groupby(['NOC','Medal']).size().unstack('Medal',fill_value=0)
g1
```

```python
## Next sort by the medals and select the corresponding columns

g2 = g1.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
g2
```
